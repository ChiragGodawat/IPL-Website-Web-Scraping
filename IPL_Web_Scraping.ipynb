```python
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
```

```python
url = "https://www.iplt20.com/stats/all-time/most-runs"
uClient = uReq(url)
html_page = uClient.read()
uClient.close()
html_page
```

```python
html_soup_page = soup(html_page)
```

```python
html_soup_page
```

```python
containers = html_soup_page.findAll("div",{"class":"top-players__player-name"})
```

```python
containers
```

```python
len(containers)
```

```python
print("Most Runs Scorer as per IPL website: ")
for container in containers:
    first_name = container.findAll("a",{"class":"top-players__player-link"})
    if(len(first_name)==0):
        print( " ".join(container.text.split()))
        continue
    print(" ".join(first_name[0].text.split()))
```
